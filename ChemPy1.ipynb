```python
def DIC_Coupling(a1, eq=2):
    #eq=2
    Ox=format(a1*1.5*eq*142.1/1000, '0.2f')
    DIC_1=format(a1*1.3*eq*156.6/1000, '0.2f')
    DIC_2=format(a1*0.7*eq*156.6/1000, '0.2f')
    print(' Oxyma (g): %s \n' %Ox,   '1st DIC (mL): % s \n' %DIC_1,  '2nd DIC  (mL): % s \n' %DIC_2)

def DIC_HOBt(a1, eq=2):
    #eq=2
    Ox=format(a1*1.5*eq*153/1000,'0.2f')
    DIC_1=format(a1*1.3*eq*156.6/1000,'0.2f')
    DIC_2=format(a1*0.7*eq*156.6/1000,'0.2f')
    print(' HOBt (g): %s \n' %Ox,   '1st DIC (mL): % s \n' %DIC_1,  '2nd DIC (mL): % s \n' %DIC_2)

def TCTU_90(a1, eq=2):
    #eq=2
    TCTU=format(a1*0.9*eq*355.5/1000,'0.2f')
    DIPEA_1=format(a1*174.2*eq/1000,'0.2f')
    DIPEA_2=format(a1*0.25*174.2*eq/1000,'0.2f')
    print(' TCTU (0.9eq) (g): %s \n' %TCTU,   'DIPEA_1 (mL): % s \n' %DIPEA_1,  'DIPEA_2 (mL): % s \n' %DIPEA_2)

def TCTU_95(a1, eq=2):
    #eq=2
    TCTU=format(a1*0.95*eq*355.5/1000,'0.2f')
    DIPEA_1=format(a1*174.2*eq/1000, '0.2f')
    DIPEA_2=format(a1*0.25*174.2*eq/1000, '0.2f')
    print(' TCTU(0.95eq) (g): %s \n' %TCTU,   'DIPEA_1 (mL): % s \n' %DIPEA_1,  'DIPEA_2 (mL): % s \n' %DIPEA_2)

def DEPBT_100(a1, eq=2):
    #eq=2
    DEPBT=format(a1*1*eq*299.2/1000,'0.2f')
    DIPEA_1=format(a1*174.2*eq/1000,'0.2f')
    DIPEA_2=format(a1*0.25*174.2*eq/1000,'0.2f')
    print(' DEPBT(1.0 eq) (g): %s \n' %DEPBT,   'DIPEA_1 (mL): % s \n' %DIPEA_1,  'DIPEA_2 (mL): % s \n' %DIPEA_2)

def DIC_Anhy(a1, eq=2):

    DIC_1=format(a1*0.5*eq*156.6/1000,'0.2f')
    #DIC_2=a1*0.7*eq*156.6/1000
    print(' DIC(Anhydide Coupling) (mL): % s \n' %DIC_1)

def Capping(a1, eq=5):
    #eq=2
    Ac=format(a1*eq*94.5/1000,'0.2f')
    DIPEA=format(a1*174.2/1000,'0.2f')
    print(' Ac2O (mL): %s \n' %Ac,   'DIPEA (mL): % s \n' %DIPEA)

def CTC_Loading(a1, a2,a3, eq=1):

    AA=format(a1*eq*a3/1000,'0.2f')
    DIPEA=format(a1*2*eq*174.2/1000, '.2f')
    print(' % s: (g)' %a2, '% s \n'  %AA,   'DIPEA (mL): % s \n' %DIPEA)

def Resin(scale,loading1=0.5, loading2=1.0):
    Amide=format(scale/loading1,'0.1f')
    CTC=format(scale/loading2,'0.2f')

    print("Ramage or Amide Resin" +' % s: (g)' %Amide)
    print("CTC Resin" +' % s: (g)' %CTC)


def prod_Scale(bb, loading1=0.5,loading2=1.0):
    Capping(bb)
    DIC_Coupling(bb,2)
    DIC_Anhy(bb)
    TCTU_90(bb)
    TCTU_95(bb)
    DEPBT_100(bb)
    aa='Fmoc-Gly-OH'
    CTC_Loading(bb, aa, 299)
    DIC_HOBt(bb)
    Resin(bb,loading1,loading2)
prod_Scale(scale)
```

```python

def Mol_wt(seq):
    totalW = 0.00
    for aa in seq:
        totalW = totalW + protweight.get(aa.upper(),0)
    totalA = totalW-(18.01*(len(seq)-1))+18
    totalB = totalW-(18.01*(len(seq)-1))+17
    print("The net weight for peptide acid is: "+format(totalA,'0.2f'))
    print("The net weight for peptide Amide is: "+format(totalB,'0.2f'))
    return totalA,totalB

print("Molecular Weight: ", Mol_wt(pp))

def charge(seq):

    charge = -0.002
    AACharge = {"C":-.045,"D":-.999,"E":-.998,"H":.091,
        "K":1,"R":1,"Y":-.001}
    for aa in seq:
        if aa in AACharge:
            charge += AACharge[aa]
        else:
            pass
    print(charge)
    return charge
print("charge",charge(pp))
```
